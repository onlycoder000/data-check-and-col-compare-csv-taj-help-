```python
import csv
import sys
```

```python
z=0
with open('FDIC Banks.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        ok=False
        with open('fl_user.csv', 'r') as file2:
            reader2 = csv.reader(file2)
            for row2 in reader2:
                domain=row2[5]
                for i in row:
                    if domain in i:
                        ok=True
                        break
                if(ok==True):
                    break
                
                    
        
        if ok:
            row[4]=row[3]
            with open('data.csv', mode='a') as open_file:
                open_writer = csv.writer(open_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                open_writer.writerow(row)
        else:
            with open('data.csv', mode='a') as open_file:
                open_writer = csv.writer(open_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                open_writer.writerow(row)
                
        print(str(z)+' '+domain+' '+str(ok))
        z+=1
         
        
```
